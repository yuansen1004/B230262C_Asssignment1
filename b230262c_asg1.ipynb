```python
ttt = ["1", "2", "3",
       "4", "5", "6",
       "7", "8", "9"]

currentPlayer = "X"
winner = None
gameRunning = True

def printTicTacToe(ttt):
    print(ttt[0] + " | " + ttt[1] + " | " + ttt[2])
    print("--+---+--")
    print(ttt[3] + " | " + ttt[4] + " | " + ttt[5])
    print("--+---+--")
    print(ttt[6] + " | " + ttt[7] + " | " + ttt[8])

def playerInput(ttt):
    while True:
        inp = input("It's " + currentPlayer + " turns. " + "Enter a number between 1~9: ")
        try:
            inp = int(inp) - 1
            if 0 <= inp <= 8 and ttt[inp] != "X" and ttt[inp] != "O":
                ttt[inp] = currentPlayer
                return
            else:
                print("Oops! That spot is already taken or invalid. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number between 1 and 9.")

def checkHorizontal(ttt):
    global winner
    if ttt[0] == ttt[1] == ttt[2] and ttt[1] == currentPlayer:
        winner = ttt[0]
        return True
    elif ttt[3] == ttt[4] == ttt[5] and ttt[3] == currentPlayer:
        winner = ttt[3]
        return True
    elif ttt[6] == ttt[7] == ttt[8] and ttt[6] == currentPlayer:
        winner = ttt[6]
        return True

def checkRow(ttt):
    global winner
    if ttt[0] == ttt[3] == ttt[6] and ttt[0] == currentPlayer:
        winner = ttt[0]
        return True
    elif ttt[1] == ttt[4] == ttt[7] and ttt[1] == currentPlayer:
        winner = ttt[1]
        return True
    elif ttt[2] == ttt[5] == ttt[8] and ttt[2] == currentPlayer:
        winner = ttt[2]
        return True

def checkDiag(ttt):
    global winner
    if ttt[0] == ttt[4] == ttt[8] and ttt[0] == currentPlayer:
        winner = ttt[0]
        return True
    elif ttt[2] == ttt[4] == ttt[6] and ttt[2] == currentPlayer:
        winner = ttt[2]
        return True

def checkTie(ttt):
    global gameRunning
    if "1" not in ttt and "2" not in ttt and "3" not in ttt and "4" not in ttt and "5" not in ttt and "6" not in ttt and "7" not in ttt and "8" not in ttt and "9" not in ttt:
        printTicTacToe(ttt)
        print("It is a tie!")
        gameRunning = False

def checkWin():
    global gameRunning
    if checkDiag(ttt) or checkHorizontal(ttt) or checkRow(ttt):
        print("The winner is " + winner)
        gameRunning = False

def switchPlayer():
    global currentPlayer
    currentPlayer = "O" if currentPlayer == "X" else "X"

while gameRunning:
    printTicTacToe(ttt)
    playerInput(ttt)
    checkWin()
    checkTie(ttt)
    switchPlayer()
```
